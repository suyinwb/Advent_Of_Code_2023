```python
import pprint
import pandas as pd
import numpy as np
import re
import time
import datetime
```

```python
# Bag only has: 12 red cubes, 13 green cubes, and 14 blue cubes

# Dictionary mapping string colours to their max quantity
colour_cubes_qty = {
    'red': 12,
    'green': 13,
    'blue': 14
}
```

```python
# Read the txt file

file_path = 'resources/day2_input.txt'

# Open the file and read its contents using readlines()
with open(file_path, 'r') as file:
    lines = file.readlines()
```

## PART 1

```python
def test_qty(text,numb):
    
    within_limit = False
    
    for word in colour_cubes_qty.keys():
        max_qty = int(colour_cubes_qty[word])

        if text.startswith(word) and numb <= max_qty:
            #print(text)
            #print(numb)
            #print('ok')
            within_limit = True
            break

        #if not within_limit:
            #print(text)
            #print(numb)
            #print('nok')
    
test_qty('green',10)
```

```python
# Function to check if all elements in the list are True
def check_all_true(lst):
    return all(lst)
```

```python
game_list = []

for line in lines:
    line = line.strip() # Strip to remove newline characters
    print(line)
    # Split the string at the colon ":" and get the substring before it
    game_num = line.split(':')[0].strip()
    #print(game_num)

    draws = line.split(':')[1].strip()
    #print(draws)
    
    split_turns = draws.split(';')
    #print(split_turns)


    
    game_check = []
    
    for turn in split_turns:
        
        turn_check = []
        
        game = turn.lstrip()
        #print(game)
        each_cube = game.split(',')
        
        for each in each_cube:
            result = None
            cube = each.strip()
            #print(cube)
            num_colour = cube.split()
            number = int(num_colour[0] if len(num_colour) > 0 else None) # Get the element at index 0
            #print(number)
            colour = num_colour[1] if len(num_colour) > 1 else None # Get the element at index 1
            #print(colour)
            
            result = test_qty(colour,number)
            #print(result)
            turn_check.append(result)
#             if result is True:

#             else:
#                 break
                
        print('This turn results is:')
        print(turn_check)
        
        if check_all_true(turn_check) is True:
            #print(f'Yes')
            game_check.append(True)
        else:
            #print(f'No')
            game_check.append(False)

        print('This full game results is:')
        print(game_check)
        
    if check_all_true(game_check) is True:
        print(f'Yes {game_num}')
        game_list.append(game_num)
    else:
        print(f'NO to {game_num}')
    
print(game_list)         
```

```python
# Extract digits from each string using regular expression
digits_list = [int(re.search(r'\d+', game_num).group()) for game_num in game_list]

print(digits_list)
```

```python
print(sum(digits_list))
```

```python
def compare_qty(text,numb):
    
    for word in colour_cubes_qty.keys():
        current_qty = int(colour_cubes_qty[word])

        if text.startswith(word) and numb > current_qty:
#             print(text)
#             print(numb)
#             print('is more')
            colour_cubes_qty[word] = numb  # Update the dictionary value for the color
    
compare_qty('green',10)
compare_qty('red',10)

print(colour_cubes_qty)
```
