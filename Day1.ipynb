## PART 1

```python
def find_digits(input_string):
    return ''.join([char for char in input_string if char.isdigit()])
```

```python
result_list = []

for line in lines:
    line = line.strip()
    result = find_digits(line)
    #print(f"{''.join(result)}")
    print(result)
    result_list.append(result)
    
print(result_list)
    
```

```python
sum_of_integers = sum(digits_list)
print(sum_of_integers)
```

```python
# Dictionary mapping string names to integers
string_to_int = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9
}
```

```python
# Sample list containing string representations of numbers
string_numbers = ['one', 'three', 'two', 'four']

# Convert string names to integers using the dictionary
int_numbers = [string_to_int[s] for s in string_numbers]

print(f"The list of integers: {int_numbers}")
```

```python
# Sample string containing representations of numbers
text_with_numbers = "eighttkbtzjz6nineeight"

# Function to find and convert string representations to integers
def find_and_convert_string_to_int(text):
    result = ""
    i = 0
    while i < len(text):
        found = False
        for word in sorted(string_to_int.keys(), key=len, reverse=True):
            print(word)
            if text[i:].startswith(word):
                result += str(string_to_int[word])
                print(result)
                i += len(word)
                found = True
                break
        if not found:
            result += text[i]
            i += 1
    return result

# Convert string names to integers using the function
int_text = find_and_convert_string_to_int(text_with_numbers)
print(f"The converted string to integers: {int_text}")
```
