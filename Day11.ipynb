```python
# Read the txt file

file_path = 'resources/day11_input.txt'

# Read the input file and store grids as list of lists
grids = []
current_grid = []

with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        if line:  # Check if the line is not empty
            current_grid.append(list(line))  # Append each line as a list of characters
        else:  # Empty line indicates the end of the current grid
            if current_grid:
                grids.append(current_grid)  # Add the completed grid to the list
                current_grid = []  # Reset for the next grid (if any)

# If there's any remaining grid at the end of the file, add it
if current_grid:
    grids.append(current_grid)

# Display each grid separately (as list of lists)
for idx, grid in enumerate(grids, start=1):
    print(f"Grid {idx}:")
    for row in grid:
        print(''.join(row))
    print()

print(grids)
```

## PART 1

```python
# Convert characters to integers: '.' to 0 and '#' to 1
for grid in grids:
    converted_list = [[0 if char == '.' else 1 for char in row] for row in grid]
    print(converted_list)

    # Convert the list of lists to a NumPy array
    arr = np.array(converted_list)
    print("2D NumPy array:")
    print(arr)
```

```python
#The script first identifies the indices of rows with only zeros, then inserts the same row below each identified row. 

rows_to_insert = []

for i, row in enumerate(arr):
    if np.array_equal(row, np.zeros_like(row)):
        rows_to_insert.append(i)

for index in sorted(rows_to_insert, reverse=True):
    arr = np.insert(arr, index+1, arr[index], axis=0)

print(arr)
```

```python
#Transpose the array above and do the same for the columns

#The script first identifies the indices of columns with only zeros, then inserts the same columns below each identified column. 

transposed_arr = arr.T

rows_to_insert = []

for i, row in enumerate(transposed_arr):
    if np.array_equal(row, np.zeros_like(row)):
        rows_to_insert.append(i)

for index in sorted(rows_to_insert, reverse=True):
    transposed_arr = np.insert(transposed_arr, index+1, transposed_arr[index], axis=0)

arr = transposed_arr.T
print(arr)
```

```python
# Convert 0 to '.' and 1 to '#' within the array
converted_arr = np.where(arr == 0, '.', '#')

# Convert the NumPy array to a list of lists
galaxy_list = converted_arr.tolist()
print(galaxy_list)
print()
# Print the resulting list of lists
for galaxy in galaxy_list:
    print(' '.join(galaxy))
```

```python
# Initialize a counter for unique numbering
counter = 1

# Iterate through the array and replace each 1 with a unique number
for i in range(arr.shape[0]): #row 
    for j in range(arr.shape[1]): #column
        if arr[i, j] == 1:
            arr[i, j] = counter
            counter += 1
print(arr)

# Find the maximum value in the array
largest_galaxy = np.max(arr)
print("Largest integer:", largest_galaxy)
```

```python
galaxies_index = []
for galaxy in range(1,largest_galaxy+1,1):
    gal_index = np.where(arr == galaxy)
    gal_index = [[i, j] for i, j in zip(gal_index[0], gal_index[1])] # keep as list in a list instead of array
    galaxies_index.append(gal_index)

short_path_arr = np.zeros((largest_galaxy,largest_galaxy), dtype=int)

for i, value in enumerate(galaxies_index):
    print(value[0])
    for j in range(0,len(galaxies_index),1):
        if short_path_arr[i][j]== 0:
            # print(value[0])
            # print(value[0][0])
            # print(value[0][1])
           # print(galaxies_index[j][0][0])
            path = abs(galaxies_index[i][0][0] - galaxies_index[j][0][0]) + abs(galaxies_index[i][0][1] - galaxies_index[j][0][1])
            short_path_arr[i][j] = path
            
    
    
print(galaxies_index)
print(short_path_arr)

# print(galaxies_index[0])
# print(galaxies_index[0][0])
# print(galaxies_index[0][0][0])
# print(galaxies_index[0][0][1])
# print(galaxies_index[1][0])
# print(galaxies_index[1][0][0])
# print(galaxies_index[1][0][1])

# print(short_path_arr)
# print(short_path_arr[0][1])
# # print(short_path_arr[0][1])
```
