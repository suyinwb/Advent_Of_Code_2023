```python
import pprint
import pandas as pd
import numpy as np
import re
import time
import datetime

from collections import Counter
```

```python
# Read the txt file

file_path = 'resources/day4_input.txt'

# Open the file and read its contents using readlines()
with open(file_path, 'r') as file:
    lines = file.readlines()
```

## PART 1

```python
won_list = {}

for line in lines:

    win_list = []
    lott_list = []

    line = line.strip() # Strip to remove newline characters
    #print(line)
    # Split the string at the colon ":" and get the substring before it
    game_num = line.split(':')[0].strip()
    print(f'Game: {game_num}')
    
    winnumb_lottnumb = line.split(':')[1].strip()
    #print(winnumb_lottnumb)
    
    winnumb = winnumb_lottnumb.split('|')[0].strip()
    print(f'Winning #: {winnumb}')
    win_list = [int(each) for each in winnumb.split()]
    print(win_list)

    lottnumb = winnumb_lottnumb.split('|')[1].strip()
    print(f'Lottery #: {lottnumb}')
    lott_list = [int(each) for each in lottnumb.split()]
    print(lott_list)
    
    times = 0
    for each in lott_list:
        if each in win_list:
            won = each
            print(won)
            times +=1
    
    won_list[game_num] = times
    print(f'Won Times in {game_num}: {times}')

print(won_list)
```

```python
# computed values
computed_list = []

for value in won_list.values():
    if value > 1:
        new_value = (pow(2,value-1))
    else:
        new_value = value
        
    print(new_value)
    computed_list.append(new_value)
```

```python
won_dict = {}

for line in lines:

    win_list = []
    lott_list = []

    line = line.strip() # Strip to remove newline characters
    #print(line)
    # Split the string at the colon ":" and get the substring before it
    game_nu = line.split(':')[0].strip()
    
    print(f'Game: {game_nu}')
    
    game_num = re.findall(r'\d+', game_nu)
    game_num = ''.join(game_num)
    
    print(f'Game: {game_num}')
    
    winnumb_lottnumb = line.split(':')[1].strip()
    #print(winnumb_lottnumb)
    
    winnumb = winnumb_lottnumb.split('|')[0].strip()
    print(f'Winning #: {winnumb}')
    win_list = [int(each) for each in winnumb.split()]
    print(win_list)

    lottnumb = winnumb_lottnumb.split('|')[1].strip()
    print(f'Lottery #: {lottnumb}')
    lott_list = [int(each) for each in lottnumb.split()]
    print(lott_list)
    
    times = 0
    for each in lott_list:
        if each in win_list:
            won = each
            print(won)
            times +=1
    
    won_dict[game_num] = times
    print(f'Won Times in {game_num}: {times}')
    
won_dict = {int(key): value for key, value in won_dict.items()}

print(won_dict)
```
