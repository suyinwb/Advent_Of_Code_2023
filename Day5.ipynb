part 1: you are given 4 seeds, and a bunch of maps each with its own lines, like this: (a, b, c)

a, b, c means: [b, b + c) is transformed to [a, a + c) (square bracket inclusive, round bracket exclusive).

50 98 2 , therefore it is 98 to 99, becomes 50 to 51: when 98 becomes 50, what's the difference? -48
52 50 48, therefore it is 50 to 97, becomes 52 to 99: when 50 becomes 52, what's the difference? +2

Firstly examine the 2 ranges set above after permutation. 
Set 1: 50 to 52
Set 2: 52 to 99

Give the new sets above, where does the seed number '79' lies?
In Set 2. Therefore 79 plus Set 2' difference (+2) will become 81.



therefore seed 79 occurs in the 

with the given example, [98, 100) is transformed to [50, 52), in other words, 98 => 50, 99 => 51, 
(note 100 is not included here).

Similarly, [50, 98) => [52, 100) means 50 => 52, 51 => 53.....97 => 99.

So the given seed 79 is turned into soil 81.

Next map:

[15, 52) => [0, 37); then [52, 54) => [37, 39); [0, 15) => [39, 54)

Since 81 is not in any of these ranges, soil 81 is turned into fertilizer 81.

Similarly, fertilizer 81 = water 81 (no transformation range for fertilizer to water includes 81).

Last example: water to light.

[18, 25) => [88, 95); [25, 95) => [18, 88). So 81 is in the second transformation, and is turned into light 74.

What you need to do: turn the 4 given seeds into soil, then from soil to fertilizer, then from fertilizer to water.
lastly from humidity to location. The minimum of the 4 result locations is the answer.

```python
# Use regular expressions to find all numbers after "map:"
matches = re.findall(r'map:\s*([\d\s]+)$', text, re.MULTILINE)
#matches = re.search(r'map:\s*([\d\s]+)\S', text, re.DOTALL)
# matches = re.findall(r'map:\s*([\d\s]+)$', text)

#Print the digits found at the end of lines
print("Digits at the end of lines:", matches)

transformed_list=[]

for seed in seeds:
    
    transformed = seed
    print(f'{transformed}')

    for each in matches:
    #     print(each)
        occurance = len(re.findall(r'\n', each))
        print(occurance)
        
        i = 0
        while i  < occurance:
            sequence = re.findall(r'(.*?)\n', each)
            print(sequence[i])

            elements = sequence[i].split()
            new_begin = int(elements[0])
            old_begin = int(elements[1])
            move_by = int(elements[2])

            old_end = old_begin + move_by - 1
            print(f'Old_end: {old_end}')
            new_end = new_begin + move_by - 1
            print(f'New End: {new_end}')
            sets_shift = new_begin - old_begin
            print(f'Shifted By: {sets_shift}')

            if old_begin <= transformed <= old_end:
                transformed = transformed + sets_shift
                print(f'New Transformed Number: {transformed}')
                break
            else:
                #transformed = transformed
                print(f'New Transformed Number: {transformed}')
                pass

            i+=1
            
    transformed_list.append(transformed)
    print('==================')

print(transformed_list)
print(min(transformed_list))
```
