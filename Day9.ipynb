```python
next_predicted = []

lines = text.splitlines()

for line in lines:
# Convert the list of strings to a list of integers
    line = list(map(int, line.split()))
    print(line)
    sequence = line

    results = []  # List to store the last element from each deduction step
    results.append(sequence[-1])
    while not all(elem == 0 for elem in sequence):
        sequence = [sequence[i + 1] - sequence[i] for i in range(len(sequence) - 1)]
        print(sequence)  # Print the intermediate steps (optional)
        results.append(sequence[-1])
        
    print(sum(results))
    next_val = sum(results)
    next_predicted.append(next_val)


print(f'Sum of all next predicted value are: {sum(next_predicted)}')
    
```

```python
next_predicted = []

for line in lines:
    # Convert the list of strings to a list of integers
    sequence = list(map(int, line.split()))
    print(sequence)

    results = []  # List to store the last element from each deduction step
    results.append(sequence[0])
    
    while not all(elem == 0 for elem in sequence):
        sequence = [sequence[i] - sequence[i - 1] for i in range(len(sequence) - 1, 0 , -1)]
        print(sequence)  # Print the intermediate steps (optional)
        results.append(sequence[0])

    print("First elements from each deduction:", results)
    print("Sum of first elements:", sum(results))
    next_val = sum(results)
    next_predicted.append(next_val)

print(f"Sum of all next predicted values: {sum(next_predicted)}")
```
